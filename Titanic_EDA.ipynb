```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

sns.set(style="whitegrid", palette="pastel")
plt.rcParams['figure.figsize'] = (8,5)
```

```python
df = pd.read_csv("Titanic-Dataset.csv")

print("ðŸ”¹ First 5 rows of the dataset:")
print(df.head())

print("\nðŸ”¹ Dataset Info:")
print(df.info())

print("\nðŸ”¹ Summary Statistics:")
print(df.describe())
```

```python
# Summary Statistics (Mean, Median, Std)

print("\nðŸ”¹ Mean of numeric columns:\n", df.mean(numeric_only=True))
print("\nðŸ”¹ Median of numeric columns:\n", df.median(numeric_only=True))
print("\nðŸ”¹ Standard Deviation of numeric columns:\n", df.std(numeric_only=True))
```

```python
# Visualize Numeric Features
numeric_cols = df.select_dtypes(include=['int64','float64']).columns
# Histograms for numeric columns
df[numeric_cols].hist(figsize=(10,8), bins=15, color='skyblue', edgecolor='black')
plt.suptitle("ðŸ“Š Histograms of Numeric Features", fontsize=14)
plt.show()

# Boxplots for numeric columns
plt.figure(figsize=(10,6))
sns.boxplot(data=df[numeric_cols])
plt.title("ðŸ“¦ Boxplot of Numeric Features")
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation & Pair Relationships

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("ðŸ”¥ Correlation Matrix")
plt.show()

# Pairplot for visual relationships
sns.pairplot(df, diag_kind='kde', corner=True)
plt.suptitle("ðŸ”— Pairplot - Feature Relationships", y=1.02)
plt.show()
```

```python
# Identify Patterns & Insights
# Survival rate by Gender
plt.figure(figsize=(6,4))
sns.countplot(x='Sex', hue='Survived', data=df, palette='cool')
plt.title("ðŸš» Survival Count by Gender")
plt.show()
```

```python
# Average Fare by Class
sns.barplot(x='Pclass', y='Fare', data=df, estimator=np.mean, palette='mako')
plt.title("ðŸ’° Average Fare by Class")
plt.show()
```
